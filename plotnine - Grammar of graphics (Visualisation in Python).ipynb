```python
#Read the data
from plotnine import *
from plotnine.data import mtcars
mtcars.head()
```

#### Visualize 2D

```python
(ggplot(mtcars,
       aes('wt','mpg'))
       +geom_point()
       +theme_bw())
```

#### Visualize 3D

```python
(ggplot(mtcars,
       aes('wt','mpg', color = 'factor(gear)'))
       +geom_point()
       +theme_bw())
```

#### Visualise 4D

```python
(ggplot(mtcars,
       aes('wt','mpg', color = 'factor(gear)', size='cyl'))
       +geom_point()
       +theme_bw())
```

```python
(ggplot(mtcars,
       aes('wt','mpg', color = 'factor(gear)', size='cyl'))
       +geom_point()
       +facet_wrap('~cyl')
       +theme_bw())
```

#### Visualize data dimensions & Statistics

```python
(ggplot(mtcars,
       aes('wt','mpg', color = 'factor(gear)', size='cyl'))
       +geom_point()
       +stat_smooth(method = 'lm')
       +theme_bw())
```

#### Visualize 5D

```python
(ggplot(mtcars,
       aes('wt','mpg', color = 'factor(gear)', size='cyl'))
       +geom_point()
       +facet_wrap('~am')
       +theme_bw())
```

#### Visualize 6D

```python
(ggplot(mtcars,
       aes('wt','mpg', color = 'factor(gear)', size='cyl'))
       +geom_point()
       +facet_grid('am ~ carb')
       +theme_bw())
```
